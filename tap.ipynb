```python
from scipy.sparse import lil_matrix
from sklearn.feature_extraction import DictVectorizer 
import math
from sklearn.cross_validation import KFold
# if has hashtag
# num of hashtag
# if has mention
# num of mention
# if has http (url include vine and video(like youtube))
# num of http
# if has images
# num of images
# if has videos (nope)
# num of videos (nope)
# if has location info
# time passed
# description
# num of followers (user)
# num of friends(followings) (user)
# ratio of followers/friends (user)
# num of status (user)
# if default_profile (user)
# if default_profile_image (user)
# num of favourites_count (user) 
# if geo_enabled (user)
# if contributors_enabled (user)
# if is_translator (user)
# lang (user) (nope)
# list count (user)
# profile_background_tile (user)
# profile_use_background_image (user)
# if protected (user)
# if is verified (user)


# ** filter by language first 
        
def ObjectToDic(obj,logTimeRate,logFollowersRate,logFriendsRate):
    dic={}
    tweet=obj['tweet']
    tweetEntites=tweet['entities']
    user=tweet['user']
    
    dic['if_hashtags']=1 if len(tweetEntites['hashtags'])!=0 else 0
    dic['hashtags_count']=len(tweetEntites['hashtags'])
    dic['if_mentions']=1 if len(tweetEntites['user_mentions'])!=0 else 0
    dic['mentions_count']=len(tweetEntites['user_mentions'])
    dic['if_urls']=1 if len(tweetEntites['urls'])!=0 else 0
    dic['urls_count']=len(tweetEntites['urls'])
    dic['if_images']=1 if 'media' in tweetEntites and len(tweetEntites['media'])!=0 else 0
    dic['images_count']=len(tweetEntites['media']) if 'media' in tweetEntites else 0
    
    dic['time_passed']=obj['time_difference_in_second'] if obj['time_difference_in_second']>0 else 0
    dic['log_time_passed']=math.log(dic['time_passed']+1.0,logTimeRate)
    
    dic['user_not_default_profile']= 0 if user['default_profile']==True else 1
    dic['user_not_default_profile_image']= 0 if user['default_profile_image']==True else 1
    dic['user_favourites_count']=user['favourites_count']
    dic['user_geo_enabled']= 1 if user['geo_enabled']==True else 0
    dic['user_contributors_enabled']= 1 if user['contributors_enabled'] else 0
    dic['user_is_translator']= 1 if user['is_translator']==True else 0
    dic['user_followers_count']=user['followers_count'] if not math.isnan(user['followers_count']) else 0
    dic['log_user_followers_count']=math.log(dic['user_followers_count']+1.0,logFollowersRate)
    dic['user_friends_count']=user['friends_count'] if not math.isnan(user['friends_count']) and user['friends_count']>0  else 0
    dic['log_user_friends_count']=math.log(dic['user_friends_count']+1.0,logFriendsRate)
    dic['user_ff_ratio']=1.0*(dic['user_followers_count']+1.0)/(dic['user_friends_count']+1.0)
    dic['log_user_ff_ratio']=1.0*(dic['log_user_followers_count']+1.0)/(dic['log_user_friends_count']+1.0)
    dic['user_listed_count']=user['listed_count'] if user['listed_count']!=None and user['listed_count']>0 else 0
    dic['log_user_listed_count']=math.log(dic['user_listed_count']+1.0,logFollowersRate)
    dic['user_profile_background_tile']= 1 if user['profile_background_tile'] else 0
    dic['user_not_profile_use_background_image']= 0 if user['profile_use_background_image'] else 1
    dic['user_protected']= 1 if user['protected'] else 0
    dic['user_statuses_count']=user['statuses_count'] if user['statuses_count']!=None and user['statuses_count']>0 else 0
    dic['log_user_statuses_count']=math.log(dic['user_statuses_count']+1.0,logFollowersRate)
    dic['user_verified']= 1 if user['verified'] else 0
    
    return dic

def toDegree(y):
    if y<1.0:
        return 0
    return (math.log(y,2))
#     return y

def getXmatrix(objlst,vec,lr1,lr2,lr3):
    X = vec.fit_transform([ObjectToDic(obj,lr1,lr2,lr3) for obj in objlst])
#     print type(X)
#     print vec.get_feature_names()
    return X.tocsr()

def getY(objlst):
    Y=[]
    for obj in objlst:
        Y.append(toDegree(obj['tweet']['retweet_count']))
    return np.array(Y)
```

```python
def plot_trends(x, y1, y2, label1, label2):
    # Plot two trends with different y-scales
    fig, ax1 = plt.subplots()
    ax1.plot(y1, 'b.-', label=label1)
    ax1.set_ylabel(label1, color='b')
    plt.xticks(range(len(x))[::20], x[::20], rotation='90')
    ax2 = ax1.twinx()
    ax2.plot(y2, 'g.-', label=label2)
    ax2.set_ylabel(label2, color='g')
    plt.show()
```

```python
def pn_value(lexicons,text):
    p=0.0
    n=0.0
    tokens=text.strip().split()
    for token in tokens:
        if token in lexicons:
            if lexicons[token]>0.0:
                p+=lexicons[token]
            else:
                n-=lexicons[token]
    return p,n
```
